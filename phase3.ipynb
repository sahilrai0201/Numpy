## Advance operations with business example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure : [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],   #Paradise Biryani
    [2, 120000, 140000, 160000, 190000],   #Beijing Bites
    [3, 200000, 230000, 260000, 300000],   #Pizza Hub
    [4, 180000, 210000, 240000, 270000],   #Burger Point
    [5, 160000, 185000, 205000, 230000]   #Chai Point
])


# Print main title
print("==== Zomato Sales Analysis ====")


# Print shape of dataset (rows, columns)
# Output: (5, 5) → 5 restaurants, 5 columns (ID + 4 years)
print("\n Sales data shape : ", sales_data.shape)


# Print first 3 restaurants (row slicing)
# 0:3 → selects rows 0, 1, 2
print("\n Sample data for first three restaurant : ", sales_data[0:3])


# Print only sales data (excluding restaurant_id)
# : → all rows
# 1: → columns from index 1 to end (2021–2024 sales)
print("\n ", sales_data[:, 1:])
```

```python
# total sales per year

print(np.sum(sales_data, axis=0))

yearly_total = np.sum(sales_data[:, 1:], axis=0)
print(yearly_total)

# minimum sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis=1)
print(min_sales)

# Maximum sales per year
max_sales = np.max(sales_data[:, 1:], axis=0)
print(max_sales)

# Average sales per restaurant
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print(avg_sales)
```

```python
# Cummulative sum of each restaurant per year
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumsum)


plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumulative sales across all restaurants ")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector addition : ", vector1 + vector2)

print("\nVector multiplication : ", vector1 * vector2)

print("\nDot product : ", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("\nAngle is : ", angle)
```
