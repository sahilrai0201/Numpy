```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg     # Used to read image as NumPy array

try:
    # Load image file into NumPy array (H x W x Channels)
    logo = mpimg.imread('numpy-logo.png')

    # Remove alpha channel (keep only RGB)
    # :3 → selects Red, Green, Blue channels only
    logo = logo[:, :, :3]   # ✅ remove alpha channel

    # Display
    # Create figure for plotting images
    plt.figure(figsize=(15, 10))

    # -------- Original Image --------
    plt.subplot(121)    # 1 row, 2 columns, position 1
    plt.imshow(logo)     # Display original image
    plt.title("NumPy logo")    # Title of first image
    plt.axis('on')  # Show axis for reference   # better than grid(False) for images

    # -------- Image Transformation (Inversion) --------
    # Invert colors: creates negative/dark effect
    dark_logo = 1 - logo

    plt.subplot(122)    # 1 row, 2 columns, position 2
    plt.imshow(dark_logo)   # Display inverted image
    plt.title("NumPy Dark logo")    # Title of second image
    plt.axis('on')  # Show axis for reference    # better than grid(False) for images

    # Display both images together
    plt.show()

# Handle case where image file is missing
except FileNotFoundError:
    print("File not found")

# Handle any other unexpected errors
except Exception as e:
    print("Error:", e)  
```
